# Line plot

```python
x  = [1,2,3,4]
y  = [2,3,4,5]
x1 = [2.1,3.2,4.3,5.4]
y1 = [3.5,4.6,5.7,6.8]
```

```python
plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')
plt.title('Line plot title')
plt.grid(False)                         # or True
plt.xlim([0,7])
plt.ylim([0,7])

plt.plot(x, y, label='x vs y')
plt.plot(x1, y1, label='x1 vs y1')

plt.legend()
```

# Scatter plot

```python
plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')
plt.title('Scatter plot title')
plt.grid(False)                         # or True
plt.xlim([0,7])
plt.ylim([0,7])

plt.scatter(x, y, label='x vs y')
plt.scatter(x1, y1, label='x1 vs y1')

plt.legend()
```

# Line plot with error bars

```python
t  = [1, 2, 3, 4, 5, 6, 7]
u  = [1, 1.5, 1, 0.7, -0.3, -0.9, 0]
du = [0.1, 0.2, 0.3, 0.4, 0.3, 0.2, 0.1]    # LIST of error bar only for u

# Another example
# import numpy as np
# t  = np.linspace(1, 7, 10)
# u  = np.sin(t) + du * np.random.rand(10)
# du = 0.7
```

```python
plt.xlabel('t')
plt.ylabel('u', rotation='horizontal')
plt.title('Line plot with error bars')
plt.grid(False)                         # or True
plt.xlim([0,8])
plt.ylim([-2,3])

plt.errorbar(t,u, yerr=du, label='t vs u w/ error bar')

plt.legend()
```

# Scatter plot with error bars

```python
plt.xlabel('t')
plt.ylabel('u', rotation='horizontal')
plt.title('Scatter plot with error bars')
plt.grid(False)                         # or True
plt.xlim([0,8])
plt.ylim([-2,3])

plt.errorbar(t,u, yerr=du, fmt='o', label='t vs u w/ error bar')

plt.legend()
```

# Polar line/scatter plot with or without error bars

```python
import numpy as np

r     = np.arange(0, 4, 0.5)
theta = 2 * np.pi * r
```

```python
x_error = 0.1
y_error = 0.1
```

```python
plt.polar(r, theta, label='r vs theta')
plt.legend(loc = 'best')
```

```python
plt.polar(r, theta, 'bo', label='r vs theta')
plt.legend(loc = 'lower left')
```

```python
plt.polar(r, theta, 'ro', label='r vs theta')
plt.errorbar(r, theta, xerr = x_error, yerr = y_error, fmt='o', color='red')
plt.legend(loc = 'upper right')
```
